```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Predict the firm's future gross profit
# expected revunce and COGs 
#Using 1000 simluation of companysnexpected revenues
```

```python
rev_m = 170 #expected revuence 
rev_stdev = 20 # standard deviation is 20 millions
iterations = 1000 # from normal distrubation 
```

```python
rev = np.random.normal(rev_m, rev_stdev,iterations) # produce future revunces using random normal disturbation 
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))
#impact of COGS is 60% of revenues normal distrunbation of 10% by the mean of Cost of good solds
#0.1 must be from a random value of standard deviation 
plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

```python
#COGS is a negative so do the inverse to get Groos _profit 
#
```

```python
Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show()
```
